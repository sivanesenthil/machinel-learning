#interpolate
1.method(linear,time,index,nearest,polynomial,spline)
2.axis
3.limit
4.limit direction
5.limit area
6.inplace

```python
#interpolate-linear
s=pd.Series([0,1,np.nan,3])
s_interpolated=s.interpolate()
print(s_interpolated)
```

```python
#polynomial
s_interpolated_poly = s.interpolate(method='polynomial',order=2)
print(s_interpolated_poly)
```

```python
#time
df=pd.DataFrame({
    'date':pd.date_range('2023-01-02',periods=6,freq='D'),
    'value':[3,np.nan,np.nan,6,7,8]
})
df.set_index('date',inplace=True)
df_interpolated = df.interpolate(method='time')
print(df)
print(df_interpolated)
```
